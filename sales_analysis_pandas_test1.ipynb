### Task 2: Add Month Column 

```python
all_data = pd.read_csv(r"C:\Users\akars\Documents\pandas_python\all_data.csv")
all_data.head()
```

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Day & Year'] = all_data['Order Date'].str[3:8]
all_data.head()
```

### Clean NaN or other useless data

```python
all_data.head(50)
```

```python
nan_df = all_data[all_data.isna().any(axis=1)]
nan_df.head()
```

```python
all_data = all_data.dropna(how='all')
all_data.head()
```

```python
all_data['Month'] = all_data['Month'].astype('int32')
```

### Find 'Or' and delete it

```python
temp_df = all_data[all_data['Order Date'].str[0:2]=='Or']
temp_df.head()
```


    ### remove the above unnessary data

```python
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
all_data.head()
```

```python
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

```python
all_data['revenue'] = 1
all_data['Quantity Ordered'] = all_data['Quantity Ordered'].astype('int32')
all_data['Price Each'] = all_data['Price Each'].astype('float')
all_data['revenue'] = all_data['revenue'].astype('float')
all_data['revenue'] = all_data['Quantity Ordered'] * all_data['Price Each']
#sorted_all_data = all_data.sort_values(by= 'Price Each')
all_data.head()
```

```python
results=all_data.groupby('Month').sum()
results
```

```python
import matplotlib.pyplot as plot

months = range(1,13)

plot.bar(months, results['revenue'])
#plot.xticks(months)
plot.ylabel('Revenue in US $')
plot.xlabel('Month number')
plot.show()
```
